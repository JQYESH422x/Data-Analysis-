### Removing Null Values : Imputrers 

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv(r"C:\Users\jayes\Downloads\googleplaystore.csv")
```

```python
df
```

```python
df.iloc[:,2:3]
```

```python
impute=SimpleImputer(missing_values=np.nan,strategy='mean')
x=impute.fit_transform(df.iloc[:,2:3])
```

```python
df.iloc[:,2:3]=x
```
