```python
### Numerical variables
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

##### Observations:

1.When rating is in between 4.5 to 4.9 --> Excellant

2.When rating is in between 4.0 to 4.4 --> Very good

3.When rating is in between 3.5 to 3.9 --> good

4.When rating is in between 2.5 to 3.4 --> average


```python
## to make it more accurate we have used paltte
```

```python
#barplot
plt.figure(figsize=(12,8))
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
plt.show()
```

##### Observation

1. not rated count is very high
2. maximum number of rating are between 2.5 to 3.5

```python
# countplot
sns.countplot(x='Rating color',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```
